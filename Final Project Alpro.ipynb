```python
mata_uang={
    "USD":1,
    "IDR":1660.90,
    "SGD":1.30,
    "EUR":0.86
}
# fungsi tersier
## konversi USD

def USDtoIDR(jumlah=None, kurs=None):
  if kurs is None:
    kurs = mata_uang['IDR']
  if jumlah is None:
    nominal_USDtoIDR = input("Masukan nominal USD: ")
    print("\n")
    try:
      jumlah = float(nominal_USDtoIDR)
    except ValueError:
      print("Input tidak valid. Masukkan angka untuk nominal USD.")
      return None
  hasil_konversi = jumlah * kurs
  print("-"*50)
  print(f"{nominal_USDtoIDR} USD = {hasil_konversi} IDR")
  print("-"*50)
  return hasil_konversi

def USDtoSGD(jumlah=None,kurs=None):
  if kurs is None:
    kurs = mata_uang['SGD']
  if jumlah is None:
    nominal_USDtoSGD = input("Masukan nominal USD: ")
    print("\n")
    try:
      jumlah = float(nominal_USDtoSGD)
    except ValueError:
      print("Input tidak valid. Silahkan masukan nominal USD.")
      return None
    hasil_konversi = jumlah * kurs
    print("-"*50)
    print(f"{nominal_USDtoSGD} USD = {hasil_konversi} SGD")
    print('-'*50)
    return hasil_konversi

def USDtoEUR(jumlah=None,kurs=None):
  if kurs is None:
    kurs = mata_uang['EUR']
  if jumlah is None:
    nominal_USDtoEUR = input("Masukan nominal USD: ")
    print("\n")
    try:
      jumlah = float(nominal_USDtoEUR)
    except ValueError:
      print("Input tidak valid. Silahkan masukan nominal USD.")
      return None
    hasil_konversi = jumlah * kurs
    print("-"*50)
    print(f"{nominal_USDtoEUR} USD = {hasil_konversi} EUR")
    print('-'*50)
    return hasil_konversi


##Konversi IDR

def IDRtoUSD(jumlah=None,kurs=None):
  if kurs is None:
    kurs = mata_uang['USD'] / mata_uang['IDR']
  if jumlah is None:
    nominal_IDRtoUSD = input("Masukan nominal IDR: ")
    print("\n")
    try:
      jumlah = float(nominal_IDRtoUSD)
    except ValueError:
      print("Input tidak valid. Silahkan masukan nominal IDR")
      return None
    hasil_konversi = jumlah * kurs
    print("-"*50)
    print(f"{nominal_IDRtoUSD} IDR = {hasil_konversi} USD")
    print("-"*50)
    return hasil_konversi

def IDRtoSGD(jumlah=None,kurs=None):
  if kurs is None:
    kurs = 1 / 12803.77
  if jumlah is None:
    nominal_IDRtoSGD = input("Masukan nominal IDR: ")
    print("\n")
    try:
      jumlah = float(nominal_IDRtoSGD)
    except ValueError:
      print("Input tidak valid. Silahkan masukan nominal IDR")
      return None
    hasil_konversi = jumlah * kurs
    print("-"*50)
    print(f"{nominal_IDRtoSGD} IDR = {hasil_konversi} SGD")
    print("-"*50)
    return hasil_konversi

def IDRtoEUR(jumlah=None,kurs=None):
  if kurs is None:
    kurs = 19306.3
  if jumlah is None:
    nominal_IDRtoEUR = input("Masukan nominal IDR: ")
    print("\n")
    try:
      jumlah = float(nominal_IDRtoEUR)
    except ValueError:
      print("Input tidak valid masukan nominal IDR")
      return None
    hasil_konversi = jumlah/kurs
    print("-"*50)
    print(f"{nominal_IDRtoEUR} IDR = {hasil_konversi} EUR")
    print("-"*50)
    return hasil_konversi

# konversi SGD

def SGDtoUSD(jumlah=None,kurs=None):
  if kurs is None:
    kurs = mata_uang['USD']/mata_uang['SGD']
  if jumlah is None:
    nominal_SGDtoUSD = input("Masukan nominal SGD: ")
    print("\n")
    try:
      jumlah = float(nominal_SGDtoUSD)
    except ValueError:
      print("Input tidak valid. Silahkan masukan nominal SGD")
      return None
    hasil_konversi = jumlah * kurs
    print("-"*50)
    print(f"{nominal_SGDtoUSD} SGD = {hasil_konversi} USD")
    print("-"*50)
    return hasil_konversi

def SGDtoIDR(jumlah=None,kurs=None):
  if kurs is None:
    kurs = 12803.77
  if jumlah is None:
    nominal_SGDtoIDR = input("Masukan nominal SGD: ")
    print("\n")
    try:
      jumlah = float(nominal_SGDtoIDR)
    except ValueError:
      print("Input tidak valid. Silahkan masukan nominal SGD")
      return None
    hasil_konversi = jumlah * kurs
    print("-"*50)
    print(f"{nominal_SGDtoIDR} SGD = {hasil_konversi} IDR")
    print("-"*50)
    return hasil_konversi

def SGDtoEUR(jumlah=None,kurs=None):
  if kurs is None:
    kurs = 1.51
  if jumlah is None:
    nominal_SGDtoEUR = input("Masukan nominal SGD: ")
    print("\n")
    try:
      jumlah = float(nominal_SGDtoEUR)
    except ValueError:
      print("Input tidak valid. Silahkan masukan nominal SGD")
      return None
    hasil_konversi = jumlah/kurs
    print("-"*50)
    print(f"{nominal_SGDtoEUR} SGD = {hasil_konversi} EUR")
    print("-"*50)
    return hasil_konversi

def EURtoUSD(jumlah=None,kurs=None):
  if kurs is None:
    kurs = mata_uang["EUR"]
  if jumlah is None:
    nominal_EURtoUSD = input("Masukan nominal EUR: ")
    print("\n")
    try:
      jumlah = float(nominal_EURtoUSD)
    except ValueError:
      print("Input tidak valid. Silahkan masukan nominal SGD")
      return None
    hasil_konversi = jumlah*kurs
    print("-"*50)
    print(f"{nominal_SGDtoEUR} SGD = {hasil_konversi} EUR")
    print("-"*50)
    return hasil_konversi



# fungsi Sekunder

def USD():
  print(f"\n{'Pilih Konversi ke mata uang apa':^50}")
  print("="*50)
  print("1.IDR")
  print("2.SGD")
  print("3.EUR")
  print("4.Kembali")
  choice_usd=input("pilih menu (1/2)= ")
  if choice_usd == '1':
    USDtoIDR()
  elif choice_usd == '2':
    USDtoSGD()
  elif choice_usd == '3':
    USDtoEUR()
  elif choice_usd == '4':
    return
  else:
    print("Tidak valid")

def IDR():
  print(f"\n{'Pilih Konversi ke mata uang apa':^50}")
  print("="*50)
  print("1.USD")
  print("2.SGD")
  print("3.EUR")
  print("4.Kembali")
  choice_idr=input("pilih menu (1/2)= ")
  if choice_idr == '1':
    IDRtoUSD()
  elif choice_idr == '2':
    IDRtoSGD()
  elif choice_idr == '3':
    IDRtoEUR()
  elif choice_idr == '4':
    return
  else:
    print("Tidak valid")

def SGD():
  print(f"\n{'Pilih Konversi ke mata uang apa':^50}")
  print("="*50)
  print("1.USD")
  print("2.IDR")
  print("3.EUR")
  print("4.Kembali")
  choice_sgd=input("pilih menu (1/2)= ")
  if choice_sgd == '1':
    SGDtoUSD()
  elif choice_sgd == '2':
    SGDtoIDR()
  elif choice_sgd == '3':
    SGDtoEUR()
  elif choice_sgd == '4':
    return
  else:
    print("Tidak valid")

def EUR():
  print(f"\n{'Pilih Konversi ke mata uang apa':^50}")
  print("="*50)
  print("1.USD")
  print("2.IDR")
  print("3.Kembali")
  choice_eur=input("pilih menu (1/2)= ")
  if choice_eur == '1':
    EURtoUSD()
  elif choice_eur == '2':
    print("Fitur ini belum ditambahkan")
  elif choice_eur == '3':
    print("Fitur ini belum ditambahkan")
    return
  else:
    print("Tidak valid")



# Menu Utama/fungsi utama
while True:
  print(f"\n{'SISTEM KONVERSI MATA UANG':^50}")
  print("="*50)
  print(f"{'Nilai Kurs Saat ini':^50}")
  print("-"*50)
  print(f"{'1 USD = 1660.90 IDR':<25}{'1 SGD = 0.77 USD':<25}",)
  print(f"{'1 USD = 1.30 SGD':<25}{'1 SGD = 12803.77 IDR':<25}")
  print("-"*50)
  print(f"{'USD = United State Dolar':<25}{'IDR = Indonesia Rupiah':<25}")
  print(f"{'SGD = Singaporean Dolar':<25}{'EUR = Euro':<25}")
  print("="*50)
  print("Pilih Mata Uang yang Akan Dikonversi: ")
  print("1. USD")
  print("2. IDR")
  print("3. SGD")
  print("4. EUR")
  print("5. Keluar")

  choice = input("Pilih Menu (1/2/3/4)= ")

  if choice == '1':
    USD()
  elif choice == '2':
    IDR()
  elif choice == '3':
    SGD()
  elif choice == '4':
    EUR()
  elif choice == '5':
    print("\nTerimakasih sudah menggunakan sistem ini")
    break
  else:
    print("Pilihan tidak valid")

```
